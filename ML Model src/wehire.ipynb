```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler , OrdinalEncoder
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import json
```

```python
# Load data from JSON
with open("dataset.json", "r") as f:
    data = json.load(f)

```

```python
# Convert JSON data to DataFrame
df = pd.DataFrame(data)

# Select relevant features and target
features = ["Job_Role", "Specific_Role", "Weighted_Average"]
target = "Expertise_Level"
```

```python
X = df[features]
y = df[target]

# Perform label encoding for categorical variable
# label_encoder = OrdinalEncoder  ()
# X["Job_Role"] = label_encoder.fit_transform(X["Job_Role"])
# X["Specific_Role"] = label_encoder.fit_transform(X["Specific_Role"])
```

```python
label_encoder = dict()
for cat in features:
    label_encoder[cat] = LabelEncoder()
    X[f'{cat}_n'] = label_encoder[cat].fit_transform(X[cat])

```

```python
# Normalize "Weighted_Average" using StandardScaler
scaler = StandardScaler()
X["Weighted_Average_n"] = scaler.fit_transform(X[["Weighted_Average"]])


Xn = X.drop(["Job_Role", "Specific_Role", "Weighted_Average"], axis='columns')
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(Xn, y, test_size=0.2, random_state=42)

```

```python
# Build the model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(3,)),  # Input layer for features
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)  # Output layer for regression
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

```

```python
# Train the model
model.fit(X_train, y_train, epochs=500, batch_size=32, verbose=1)
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
r2 = r2_score(y_test, y_pred)

print("R-squared:", r2)
```
